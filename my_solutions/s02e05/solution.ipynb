```python
import requests
import base64

def encode_image(image_url):
    response = requests.get(image_url, stream=True)
    if response.status_code == 200:
        image_data = response.content
        base64_image = base64.b64encode(image_data).decode('utf-8')
        return base64_image
    else:
        raise ValueError(f"Nie udało się pobrać obrazu z {image_url}. Status kod: {response.status_code}")
    
# Funkcja do generowania opisu obrazu
def describe_image(image_url):
    base64_image = encode_image(image_url)
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
        {
            "role": "user",
            "content": [
            {
                "type": "text",
                "text": "Opisz co widzisz na zdjęciu"
            },
            {
                "type": "image_url",
                "image_url": {
                "url":  f"data:image/jpeg;base64,{base64_image}"
                },
            },
            ],
        }
        ],
    )
    answer = response.choices[0].message.content
    return answer


# Funkcja do transkrypcji nagrania
def transcribe_audio(audio_url):
    response = requests.get(audio_url, stream=True)
    temp_audio_path = "temp_audio.mp3"
    
    temp_audio_path = "temp_audio.mp3"
    with open(temp_audio_path, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)
    
    # Przekaż plik do Whisper API
    with open(temp_audio_path, 'rb') as audio_file:
        transcription = client.audio.transcriptions.create(
            model="whisper-1",
            file=audio_file
        )
    return transcription

# Funkcja do przetwarzania pliku Markdown
def update_markdown(input_markdown_path, output_markdown_path):
    updated_content = ""
    with open(input_markdown_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
    
    for line in lines:
        if line.startswith("![Brak opisu]"):  # Jeśli brak opisu obrazu
            image_url = line.split('(')[-1].strip(')\n')
            print(f"Przetwarzanie obrazu: {image_url}")
            caption = describe_image(image_url)
            updated_content += f"![{caption}]({image_url})\n"
        elif line.startswith("- **rafal_dyktafon.mp3**"):  # Jeśli brak transkrypcji audio
            audio_url = line.split('(')[-1].strip(')\n')
            if audio_url != "None":
                print(f"Przetwarzanie audio: {audio_url}")
                transcription = transcribe_audio(audio_url)
                updated_content += f"- **Transkrypcja:** {transcription}\n"
                updated_content += f"  [Odtwórz nagranie]({audio_url})\n"
        else:
            updated_content += line  # Przepisz pozostałe linie bez zmian

    # Zapisz zaktualizowany plik Markdown
    with open(output_markdown_path, 'w', encoding='utf-8') as file:
        file.write(updated_content)
    print(f"Zaktualizowany plik zapisano jako: {output_markdown_path}")
```

```python
def split_long_text_into_chunks(long_text, max_tokens=1000):
    """
    Dzieli długi tekst na fragmenty, aby każdy mieścił się w podanym limicie tokenów.
    """
    words = long_text.split()
    chunks = []
    current_chunk = []

    for word in words:
        current_chunk.append(word)
        # Jeśli długość aktualnego fragmentu przekracza limit tokenów, dodaj go do wyników
        if len(current_chunk) >= max_tokens:
            chunks.append(" ".join(current_chunk))
            current_chunk = []
    
    # Dodaj ostatni fragment, jeśli nie jest pusty
    if current_chunk:
        chunks.append(" ".join(current_chunk))
    
    return chunks
```

```python
def find_relevant_chunks(chunks, question):
    """
    Wysyła fragmenty do GPT-4 i zwraca te, które są istotne dla pytania.
    """
    relevant_chunks = []

    for chunk in chunks:
        messages = [
            {
                "role": "system",
                "content": "Jesteś asystentem, który ma znaleźć fragmenty tekstu odpowiadające na pytanie użytkownika."
            },
            {
                "role": "system",
                "content": f"Fragment tekstu: {chunk}"
            },
            {
                "role": "user",
                "content": f"Czy ten fragment zawiera odpowiedź na pytanie: {question}? Jeśli tak, zwróć 'TAK'. Jeśli nie, zwróć 'NIE'."
            }
        ]

        response = client.chat.completions.create(
            model="gpt-4",
            messages=messages
        )

        answer = response.choices[0].message.content.strip()
        if "TAK" in answer:
            relevant_chunks.append(chunk)

    return relevant_chunks
```

```python
def ask_gpt_with_relevant_chunks(relevant_chunks, question):
    """
    Przekazuje wybrane fragmenty do GPT-4 i uzyskuje odpowiedź na pytanie.
    """
    combined_context = "\n\n".join(relevant_chunks)
    messages = [
        {
            "role": "system",
            "content": "Jesteś asystentem, który odpowiada na pytania użytkownika na podstawie dostarczonego kontekstu."
        },
        {
            "role": "system",
            "content": f"Kontekst: {combined_context}"
        },
        {
            "role": "user",
            "content": question
        }
    ]

    response = client.chat.completions.create(
        model="gpt-4",
        messages=messages
    )

    return response.choices[0].message.content
```

```python
def ask_question_from_long_text(long_text, question, max_tokens=1000):
    # Podziel tekst na fragmenty
    chunks = split_long_text_into_chunks(long_text, max_tokens)
    
    # Znajdź istotne fragmenty
    relevant_chunks = find_relevant_chunks(chunks, question)
    
    if not relevant_chunks:
        return "Nie znaleziono odpowiedzi na pytanie w podanym tekście."

    # Uzyskaj odpowiedź na pytanie na podstawie istotnych fragmentów
    answer = ask_gpt_with_relevant_chunks(relevant_chunks, question)
    return answer
```

```python
def split_markdown_by_section(markdown_path):
    """
    Podziel Markdown na sekcje według nagłówków.
    """
    with open(markdown_path, 'r', encoding='utf-8') as file:
        content = file.read()

    sections = {}
    current_section = None
    for line in content.splitlines():
        if line.startswith("# "):  # Główne nagłówki
            current_section = line.strip()
            sections[current_section] = []
        elif current_section is not None:
            sections[current_section].append(line)

    # Połącz linie w każdej sekcji
    for section in sections:
        sections[section] = "\n".join(sections[section])

    return sections
```

```python
def split_text_into_chunks(text, max_tokens=2000):
    """
    Dzieli tekst na mniejsze fragmenty, aby każdy mieścił się w limicie tokenów.
    """
    words = text.split()
    chunks = []
    current_chunk = []

    for word in words:
        current_chunk.append(word)
        if len(current_chunk) >= max_tokens:
            chunks.append(" ".join(current_chunk))
            current_chunk = []

    if current_chunk:
        chunks.append(" ".join(current_chunk))
    
    return chunks
```

```python
import openai

# openai.api_key = "YOUR_API_KEY"

def ask_gpt(question, context):
    """
    Wysyła pytanie do GPT-4 z podanym kontekstem.
    """
    messages = [
        {
            "role": "system",
            "content": "Jesteś asystentem, który odpowiada na pytania użytkownika na podstawie dostarczonego kontekstu."
        },
        {
            "role": "system",
            "content": f"Kontekst: {context}"
        },
        {
            "role": "user",
            "content": question
        }
    ]

    response = client.chat.completions.create(
        model="gpt-4",
        messages=messages
    )

    return response.choices[0].message.content
```

```python
def ask_question_from_markdown(markdown_path, question):
    # Podziel Markdown na sekcje
    sections = split_markdown_by_section(markdown_path)

    # Wybierz sekcję "Tekst Główny" i podziel ją na fragmenty
    if "## Tekst Główny" in sections:
        main_text = sections["## Tekst Główny"]
        chunks = split_text_into_chunks(main_text)
    else:
        chunks = []

    # Przeszukaj inne sekcje (Obrazy, Nagrania Audio)
    context = ""
    for section, content in sections.items():
        if "Obrazy" in section or "Nagrania Audio" in section:
            context += content + "\n\n"

    # Połącz wszystkie istotne części w jeden kontekst
    context += "\n".join(chunks)

    # Zapytaj GPT
    answer = ask_gpt(question, context)
    return answer
```

```python
def ask_specific_questions(markdown_path, questions):
    """
    Odpowiada na konkretne pytania na podstawie Markdowna.
    """
    # Podziel Markdown na sekcje
    sections = split_markdown_by_section(markdown_path)
    
    # Połącz cały kontekst z sekcji dla analizy
    full_context = "\n\n".join(sections.values())

    answers = {}
    
    # Odpowiadaj na każde pytanie
    for question_id, question_text in questions.items():
        # Wyślij zapytanie do GPT z pełnym kontekstem
        answer = ask_gpt(question_text, full_context)
        answers[question_id] = answer.strip()
    
    return answers

# Pytania użytkownika
questions = {
    "01": "Jakiego owocu użyto podczas pierwszej próby transmisji materii w czasie?",
    "02": "Na rynku którego miasta wykonano testową fotografię użytą podczas testu przesyłania multimediów?",
    "03": "Co Bomba chciał znaleźć w Grudziądzu?",
    "04": "Resztki jakiego dania zostały pozostawione przez Rafała?",
    "05": "Od czego pochodzą litery BNW w nazwie nowego modelu językowego?"
}

# Ścieżka do pliku Markdown
markdown_path = "updated_markdown_with_images.md"

# Uzyskaj odpowiedzi na pytania
answers = ask_specific_questions(markdown_path, questions)

# Wyświetl odpowiedzi
for question_id, answer in answers.items():
    print(f"Pytanie {question_id}: {questions[question_id]}")
    print(f"Odpowiedź: {answer}\n")
```
